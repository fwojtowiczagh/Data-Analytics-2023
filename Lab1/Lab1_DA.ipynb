```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
uniform_dist_arr = np.random.uniform(0, 1, 1000)
norm_dist_std_1_arr = np.random.normal(0, 1, 1000)
norm_dist_std_10_arr = np.random.normal(0, 10, 1000)
beta_dist_arr = np.random.beta([1, 2, 3], [1, 2, 9], (1000, 3))  # there I generate beta distribution for all three pairs
# beta_array_11 = np.random.beta(1, 1, 1000)
# beta_array_22 = np.random.beta(2, 2, 1000)
# beta_array_39 = np.random.beta(3, 9, 1000)

sigma1 = np.array([[1, 0], [0, 1]])
sigma2 = np.array([[2, 1], [1, 2]])
mean = np.array([0, 0])

norm_x1, norm_y1 = np.random.multivariate_normal(mean, sigma1, 10000).T
norm_x2, norm_y2 = np.random.multivariate_normal(mean, sigma2, 10000).T
```

Exercise 5

```python
# Generating all compulsory variables
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = abs(np.random.normal(0, 1, 1000))
mu = alpha + beta*x
```

```python
y = np.random.normal(mu, sigma, 1000)  # calculating final values
```

Exercise 6

```python
x_lines = np.linspace(-6, 6, 1000)
x_lines = x_lines.reshape((-1, 1000))

alpha = alpha.reshape((-1, 1000))
beta = beta.reshape((-1, 1000))

y_lines = alpha.T + beta.T@x_lines
```

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 9))

ax[0].scatter(x, y)
ax[0].set_title("y as a function of x")

ax[1].scatter(x, y)
ax[1].plot(x_lines[0], y_lines.T, color='r', alpha=0.2)
ax[1].set_title(r"y as a function of x with y = $\alpha + y\beta$")

plt.show()
```
